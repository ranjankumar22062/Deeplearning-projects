```python
#converting RGB image to grey scale image
import cv2
img=cv2.imread('/content/dog image.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA==')
type(img)
```

```python
grayscale_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
type(grayscale_image)
```
